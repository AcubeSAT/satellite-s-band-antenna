```python
# Plotting libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.pyplot import vlines

# Numerical and array operations
import numpy as np

# Format the output
from IPython.display import display, Markdown, Latex
```

```python
gainVSfreq = np.loadtxt('y_feed_7_66_data/gainVSfreq.txt', skiprows = 2)
gain = np.asarray([g[1] for g in gainVSfreq])
frequency = np.asarray([f[0] for f in gainVSfreq])
```

```python
# Plot results 

# Create the figure and the space
fig = plt.figure(figsize=(18, 6))
ax = fig.add_subplot(1, 2, 1)

# Plot the pattern
plot = ax.plot(frequency, gain)

ax.set_xlabel('Frequency [GHz]')
ax.set_ylabel('Gain [dB]')
ax.set_xlim(1, 3)
ax.grid(True, ls = '--', alpha = 0.5)
```

```python
ARvsFREQ = np.loadtxt('y_feed_7_66_data/ARvsFREQ.txt', skiprows = 2)
AR = np.asarray([a[1] for a in ARvsFREQ])
frequency = np.asarray([f[0] for f in ARvsFREQ])
```

```python
# Plot results 

# Create the figure and the space
fig = plt.figure(figsize=(18, 6))
ax = fig.add_subplot(1, 2, 1)

# Plot the pattern
ax.plot(frequency, AR)

ax.set_xlabel('Frequency [GHz]')
ax.set_ylabel('Axial Ratio [dB]')
ax.set_xlim(2.4, 2.45)
ax.grid(True, ls = '--', alpha = 0.5)
```

```python
s11 = np.loadtxt('y_feed_7_66_data/s11_2430.txt', skiprows = 2)
s11_2430 = np.asarray([s[1] for s in s11])
frequency = np.asarray([f[0] for f in s11])
```

```python
s11_10 = - 10*np.ones(frequency.shape[0])
freq_10 = np.where(abs(s11_2430 + 10) < 0.1)[0]
```

```python
# Plot results 

# Create the figure and the space
fig = plt.figure(figsize=(18, 6))
ax = fig.add_subplot(1, 2, 1)

# Plot the pattern
ax.plot(frequency, s11_2430)
ax.plot(frequency, s11_10)
ax.vlines(frequency[freq_10[0]], -30, 0, linestyles='dashed')
ax.vlines(frequency[freq_10[1]], -30, 0, linestyles='dashed')

print(freq_10[0])

ax.set_xlabel('Frequency [GHz]')
ax.set_ylabel('$S_{11}$ [dB]')
ax.set_xlim(1, 3)
ax.set_ylim(-25, 0)
ax.grid(True, ls = '--', alpha = 0.5)

plt.show()
```
